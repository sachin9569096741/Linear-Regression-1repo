```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn import linear_model
```

```python
df4 = pd.read_csv("c:\\Users\\SACHIN\\Downloads\\canada_per_capita_income.csv")
df4
```

```python
%matplotlib inline
plt.xlabel("year")
plt.ylabel("per_capita_income(US$)")
plt.scatter(df4.year, df4.per_capita_income, color= "red", marker= ".")
```

```python
reg = linear_model.LinearRegression()
reg.fit(df4[["year"]], df4.per_capita_income)
```

```python
reg.predict([[2016]])
```

```python
%matplotlib inline
plt.xlabel("year")
plt.ylabel("per_capita_income(US$)")
plt.scatter(df4.year, df4.per_capita_income, color= "red", marker= ".")
plt.plot(df4.year,reg.predict(df4[["year"]]), color="blue")
```
